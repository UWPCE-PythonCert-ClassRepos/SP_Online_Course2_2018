# Load data

```python
import pandas as pd
# music = pd.read_csv("/media/sweerts/Data/UW/SP_Online_Course2_2018/students/Jeroen_Sweerts/Lesson1/featuresdf.csv")
music = pd.read_csv("featuresdf.csv")
display(music)
```

# Print songs of favorite artist

```python
def fav_artist():
    row = 0
    def filter_song():
        nonlocal row
        artist = music.iloc[row]        
        row += 1
        return artist
    return filter_song
        
        
my_fav_artist = fav_artist()

for row in music.iterrows():
    print(my_fav_artist())
        
```
