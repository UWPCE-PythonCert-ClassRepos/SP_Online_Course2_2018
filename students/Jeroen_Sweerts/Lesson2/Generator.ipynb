```python
import pandas as pd
# music = pd.read_csv("/media/sweerts/Data/UW/SP_Online_Course2_2018/students/Jeroen_Sweerts/Lesson1/featuresdf.csv")
music = pd.read_csv("featuresdf.csv")
display(music)
```

# Print songs of favorite artist

```python
def fav_artist(artist_name):
    value = 0
    while True:
        try:
            if music.iloc[value].artists == artist_name:
                yield music.iloc[value]['name']
        except IndexError:
            break
        value = value + 1
        
artist = fav_artist("Martin Garrix")

for row in music.iterrows():
    try:
        print(next(artist))
    except StopIteration:
        break
        
```
